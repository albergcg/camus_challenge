```python
fig_size = plt.rcParams["figure.figsize"]

fig_size[0] = 7
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size
```

```python
train_path = "../data/raw/train/2ch_data"
```

```python
def mhd_to_array(path):
    return sitk.GetArrayFromImage(sitk.ReadImage(path, sitk.sitkFloat32))
```

```python
image = sitk.GetArrayFromImage(sitk.ReadImage(train_path + "/frames/patient0001_2CH_ED.mhd", sitk.sitkFloat32))
mask = sitk.GetArrayFromImage(sitk.ReadImage(train_path + "/masks/patient0001_2CH_ED_gt.mhd", sitk.sitkFloat32))
```

```python
np.shape(image)
```

```python
np.shape(mask)
```

```python
fig, ax = plt.subplots(2)
ax[0].imshow(image[0]) # End Diastolic frame
ax[1].imshow(mask[0]) # End Diastolic mask
plt.show()
```

```python
x = mhd_to_array(train_path + "/frames/patient0005_2CH_ED.mhd").shape[2]
x
```

```python
widths = list()
lengths = list()
patient_list = os.listdir("../data/raw/train/2ch_data/frames/")

for i in patient_list:
    if "mhd" in i:
        w = mhd_to_array(train_path + "/frames/" + i).shape[2]
        l = mhd_to_array(train_path + "/frames/" + i).shape[1]
        
        widths.append(w)
        lengths.append(l)
```

```python
widhts_array = np.array(widths)
lengths_array = np.array(lengths)

plt.hist(widhts_array, bins = 15)
plt.show()
plt.hist(lengths_array, bins = 15)
plt.show()
```
